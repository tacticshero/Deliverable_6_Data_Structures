```python
import pandas as pd
```

```python
def odds(n):
  odds_list = []
  for nth in range(0,n):
    odd = nth*2+1
    odds_list.append(odd)
  return odds_list
```

```python
def evens(n):
  evens_list = []
  for nth in range(0,n):
    even = nth*2
    evens_list.append(even)
  return evens_list
```

```python
def fib(n):
  fib=[0,1]
  if (n == 1):
    fib.pop()
  for nth in range(0,n-2):
    val = fib[-1] + fib[-2]
    fib.append(val)
  return fib
```

```python
try:
  length = int(input("enter an integer value greater than 0: "))
except ValueError:
  print("ERROR: Could not convert data to an integer.")
if (length <= 0):
  print("ERROR: integer must be greater than 0")
else:
  odds_list = odds(length)
  evens_list = evens(length)
  fib_list = fib(length)

  df = pd.DataFrame({'Even': evens_list, 'Odd': odds_list, 'Fibonacci': fib_list})

df
```
